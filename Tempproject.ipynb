```python
import matplotlib.pyplot as plt
import random

days = list(range(1, 31))
temperatures = [random.randint(20, 40) for _ in range(30)]

max_temp = max(temperatures)
min_temp = min(temperatures)

max_day = days[temperatures.index(max_temp)]
min_day = days[temperatures.index(min_temp)]

plt.figure(figsize=(10, 5))
plt.plot(days, temperatures, marker='o', color='orange')

plt.scatter(max_day, max_temp, color='red', s=120, label='Hottest Day')
plt.scatter(min_day, min_temp, color='blue', s=120, label='Coldest Day')

plt.annotate(f'Hottest: {max_temp}C',
             xy=(max_day, max_temp),
             xytext=(max_day, max_temp + 2),
             arrowprops=dict(facecolor='red'))

plt.annotate(f'Coldest: {min_temp}C',
             xy=(min_day, min_temp),
             xytext=(min_day, min_temp - 4),
             arrowprops=dict(facecolor='blue'))

plt.title("Daily Temperature Over a Month")
plt.xlabel("Day")
plt.ylabel("Temperature (C)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
